**15.  What is causation? Explain difference between correlation and causation with an example?**

  -  Causation indicates that one event directly causes another, meaning a change in one variable is responsible for a change in another (cause-and-effect). Correlation only implies that two variables move together, not that one causes the other.

     example:

     Correlation: Two things appear to happen together. For example, ice cream sales and shark attacks both increase during summer.


     Causation: One event makes the other happen. For example, the high summer temperatures (the real cause) cause both increased ice cream sales and more people swimming, which leads to more shark attacks.

**16.  What is an Optimizer? What are different types of optimizers? Explain each with an example?**

 -  An optimizer is an algorithm used to adjust the parameters (weights and biases) of a machine learning model during training to minimize the loss function, which measures the error between the model's predictions and the actual data.

     Types of Optimizers:

  1.Gradient Descent (GD):

  example:

  In a simple linear regression problem, GD would calculate how much the total error (Mean Squared Error) changes with respect to each weight and bias in the model and adjust them iteratively until the error is minimized.

2. Stochastic Gradient Descent (SGD):

  example:

   When training a large image classification model, using the entire dataset for each update (as in GD) would be computationally expensive. SGD addresses this by randomly selecting one image at a time (or a small batch) to perform a faster, albeit noisier, update

  3.  Momentum:

  example:

  If the loss function has a steep slope in one direction and a shallow slope in another, the momentum helps the optimizer quickly traverse the shallow region without getting stuck, while reducing oscillations in the steep direction.
